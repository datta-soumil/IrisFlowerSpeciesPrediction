# IRIS DATASET CLASSIFICATION 

## Importing the Modules

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Loading The Data


```python
iris = pd.read_csv('./Iris.csv')
iris.head()
```

```python
iris.describe()
```

```python
iris.isnull().sum()
# This tells us if we have any null values in the dataset or not
```

```python
# Tells the count of each species in dataset
iris['Species'].value_counts()
```

## Removing Unneeded Column

```python
iris.drop("Id", axis=1, inplace = True)
iris.head()
```

## Performing Some EDA

```python
# Sepal Length vs Width
fig = iris[iris.Species == 'Iris-setosa'].plot(kind='scatter', x='SepalLengthCm', y='SepalWidthCm', color='orange', label='Setosa')
iris[iris.Species == 'Iris-versicolor'].plot(kind='scatter', x='SepalLengthCm', y='SepalWidthCm', color='blue', label='Versicolor', ax=fig)
iris[iris.Species == 'Iris-virginica'].plot(kind='scatter', x='SepalLengthCm', y='SepalWidthCm', color='green', label='Virginica', ax=fig)

fig.set_xlabel('Sepal Length')
fig.set_ylabel('Sepal Width')
fig.set_title('Sepal Length Vs Width')

fig=plt.gcf()
fig.set_size_inches(7, 4)
plt.show()
```

```python
# Petal Length vs Width
fig = iris[iris.Species == 'Iris-setosa'].plot(kind='scatter', x='PetalLengthCm', y='PetalWidthCm', color='orange', label='Setosa')
iris[iris.Species == 'Iris-versicolor'].plot(kind='scatter', x='PetalLengthCm', y='PetalWidthCm', color='blue', label='Versicolor', ax=fig)
iris[iris.Species == 'Iris-virginica'].plot(kind='scatter', x='PetalLengthCm', y='PetalWidthCm', color='green', label='Virginica', ax=fig)

fig.set_xlabel('Petal Length')
fig.set_ylabel('Petal Width')
fig.set_title('Petal Length Vs Width')

fig=plt.gcf()
fig.set_size_inches(7, 4)
plt.show()
```

```python
iris.hist(edgecolor='black', linewidth=1.2)
fig = plt.gcf()
fig.set_size_inches(7,4)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
sns.violinplot(x='Species', y = 'SepalLengthCm', data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='Species', y = 'SepalWidthCm', data=iris)

plt.subplot(2,2,3)
sns.violinplot(x='Species', y = 'PetalLengthCm', data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='Species', y = 'PetalWidthCm', data=iris)
```

## Correlation Matrix

```python
iris.corr()
```

```python
# Plotting the heatmap for better visualization
corr = iris.corr()
fig,ax = plt.subplots(figsize = (7,4))
sns.heatmap(corr, annot= True, cmap='cubehelix_r')
```

From this, we clearly understand that the Sepal Width and Length are not correlated whereas the Petal Width and Length are highly correlated.

## Training and Testing with various algorithms

```python
from sklearn.linear_model import LogisticRegression # for logistic regression algorithm
from sklearn.tree import DecisionTreeClassifier     # for using DTA
from sklearn.neighbors import KNeighborsClassifier  # for KNN classifier
from sklearn import svm                             # for support vector machine algorithm

from sklearn.model_selection import train_test_split # To split the dataset for training and testing 
from sklearn import metrics                          # To check the accuracy of the model
import pickle
```

```python
# Input Data
X = iris.drop(columns = ['Species'])

# Output Data
Y = iris['Species']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.30)
```

```python
# Logistic Regression 
model = LogisticRegression()
model1 = svm.SVC()
model2 = DecisionTreeClassifier()
model3 = KNeighborsClassifier(n_neighbors=3)

model.fit(x_train, y_train)
model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
model3.fit(x_train,y_train)


print('Logistic Regression Acuuracy : ', model.score(x_test, y_test) * 100)
print('Logistic Regression Acuuracy : ', model1.score(x_test, y_test) * 100)
print('Logistic Regression Acuuracy : ', model2.score(x_test, y_test) * 100)
print('Logistic Regression Acuuracy : ', model3.score(x_test, y_test) * 100)
```

```python
# Training the Model
model.fit(x_train.values, y_train.values)
```

```python
print('Acuuracy : ', model.score(x_test, y_test) * 100)
```

```python
model = svm.SVC()
model.fit(x_train, y_train)
print('The accuracy of the SVM is: ', model.score(x_test, y_test) * 100)
```
